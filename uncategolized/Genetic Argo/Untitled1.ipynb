```python

import pandas as pd
import yfinance as yf
import datetime as dt
import matplotlib.pyplot as plt
import mplfinance as mpl
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import codecs
'''
戦略
シグナル
1.25日標準偏差が7以下

買い
シグナル発生後7日以内に80日高値を更新
再び買いシグナルが出たら買い増し

売り
損切り　6%
ストップは順次あげていく
'''

#install symbols nasdaq100の銘柄コード
df = pd.read_csv('../symbols/nasdaq100.csv', encoding='shift_jis')


signs = []
for symbol in df['symbol']:
    signs.append(symbol)

    
signs = ['4477.T','6976.T','6036.T']
totalresult = []

def main():
    stock_record_list = stock_record(start = '2020-01-01' ,end = '2021-02-02')
    
    trade_record(stock_record_list)




#株式の時系列データを書く
def stock_record(start = '2020-01-01' ,end = '2021-02-02'):
    records = []

    for sign in signs:
        df2 = yf.download(sign, start, end, interval='1d') 
        records.append(df2)
        if df2.empty:
            print('empty')
            results = []
            resultNum = 7
            for num in range(resultNum):
                results.append(np.nan)
            totalresult.append(results)
            continue;
    return records


def trade_record(stock_record):
    for df in stock_record:
        #過去term日間の最高値
        term=81
        df['Highest'+str(term)] = df.iloc[:, 4].rolling(term).max()
        #標準偏差
        df['STD'] = df['Adj Close'].rolling(10).std()



if __name__ == "__main__":
    main()
```
